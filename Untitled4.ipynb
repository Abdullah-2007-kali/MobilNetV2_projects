```python
# ==========================
# 1️⃣ استيراد المكتبات الأساسية
# ==========================
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.optimizers import Adam
from sklearn.utils.class_weight import compute_class_weight
from tensorflow.keras.preprocessing import image

print("All libraries imported successfully")
```

```python
import os
import shutil

# مسار البيانات في Google Drive
drive_path = "/content/drive/MyDrive/SkinDisease_split"

# مسار البيانات المحلي في Colab
local_path = "/content/dataset"

# لو المسار المحلي مش موجود، انسخ البيانات من Google Drive
if not os.path.exists(local_path):
    shutil.copytree(drive_path, local_path)

# ==========================
# 2️⃣ تحديد مسارات البيانات
# ==========================
train_dir = "/content/dataset/train"  # مجلد بيانات التدريب
val_dir   = "/content/dataset/val"    # مجلد بيانات التحقق

# عرض عدد الصور لكل فئة داخل مجلد التدريب
classes = os.listdir(train_dir)  # الحصول على أسماء الفئات
for folder in classes:
    folder_path = os.path.join(train_dir, folder)
    print(f"{folder}: {len(os.listdir(folder_path))}")
```

```python
# ==========================
# 3️⃣ Data Augmentation وتحضير البيانات
# ==========================
IMG_SIZE = (224, 224)  # حجم الصور المطلوب لـ ResNet50
BATCH_SIZE = 32        # حجم الدفعة لكل تحديث

# Data Augmentation لزيادة تنوع البيانات وتقليل Overfitting
train_datagen = ImageDataGenerator(
    rescale=1./255,          # تطبيع قيم البكسل من 0-255 إلى 0-1
    rotation_range=20,        # تدوير الصور بشكل عشوائي ±20 درجة
    width_shift_range=0.2,    # تحريك الصورة أفقيًا بنسبة 20%
    height_shift_range=0.2,   # تحريك الصورة عموديًا بنسبة 20%
    shear_range=0.2,          # تشويه مائل عشوائي
    zoom_range=0.2,           # تكبير/تصغير عشوائي
    horizontal_flip=True,     # قلب الصور أفقيًا عشوائيًا
    fill_mode='nearest'       # طريقة ملء الفراغات بعد التحويل
)

# بيانات التحقق فقط نطبق عليها التطبيع
val_datagen = ImageDataGenerator(rescale=1./255)

# تحميل بيانات التدريب من المجلد
train_generator = train_datagen.flow_from_directory(
    train_dir,             # مسار مجلد التدريب
    target_size=IMG_SIZE,  # تغيير حجم كل صورة
    batch_size=BATCH_SIZE, # عدد الصور في كل دفعة
    class_mode='categorical'  # لأن لدينا أكثر من فئة واحدة (multi-class)
)

# تحميل بيانات التحقق من المجلد
val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)
```

```python
# ==========================
# 4️⃣ حساب class_weight للفئات غير المتوازنة
# ==========================
# إنشاء قائمة بفئات كل صورة في التدريب
y_train_classes = []
for folder in classes:
    count = len(os.listdir(os.path.join(train_dir, folder)))
    y_train_classes.extend([folder]*count)  # إضافة اسم الفئة عدد الصور

y_train_classes = np.array(y_train_classes)  # تحويل القائمة إلى مصفوفة
class_indices = train_generator.class_indices   # قاموس الفئة : رقم
y_train_labels = np.array([class_indices[y] for y in y_train_classes])
# تحويل أسماء الفئات إلى أرقام

# حساب أوزان الفئات تلقائيًا
class_weights = compute_class_weight(class_weight='balanced',
                                     classes=np.unique(y_train_labels),
                                     y=y_train_labels)
class_weights = dict(enumerate(class_weights))  # تحويلها إلى قاموس
print("Class weights:", class_weights)  # طباعة الأوزان
```

```python
# ==========================
# 5️⃣ بناء نموذج MobileNetV2مع Transfer Learning
# ==========================
# تحميل MobileNetV2 مسبق التدريب على ImageNet بدون الطبقة الأخيرة
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False  # تجميد جميع الطبقات الأساسية أولاً

# إضافة طبقات مخصصة للمشروع
x = base_model.output
x = GlobalAveragePooling2D()(x)       # تحويل المخرجات إلى مصفوفة 1D
x = Dense(128, activation='relu')(x) # طبقة كثيفة 1024 وحدة مع ReLU
x = Dropout(0.5)(x)                   # Dropout لمنع Overfitting
predictions = Dense(len(classes), activation='softmax')(x)  # طبقة الإخراج بعدد الفئات

# إنشاء النموذج النهائي
model = Model(inputs=base_model.input, outputs=predictions)

# تجميع النموذج للتحضير للتدريب
model.compile(optimizer=Adam(learning_rate=0.0001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()  # عرض تفاصيل النموذج
```

```python
# ==========================
# 6️⃣ إعداد Callbacks لإدارة التدريب
# ==========================
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau

# 1️⃣ ModelCheckpoint → لحفظ أفضل نسخة من النموذج
checkpoint = ModelCheckpoint(
    'best_skin_model_1_Dremax.keras',  # اسم الملف اللي هيُحفظ فيه النموذج
    monitor='val_accuracy',    # المقياس اللي هنراقبه (هنا خسارة التحقق)
    verbose=1,             # إظهار رسالة عند كل حفظ
    save_best_only=True,   # حفظ أفضل نسخة فقط (لما val_loss تتحسن)
    mode='max'             # "min" يعني الأقل أفضل (loss أصغر = أفضل)
)

# --------------------------------------

# 2️⃣ EarlyStopping → إيقاف التدريب المبكر لتجنب الـ overfitting
earlystop = EarlyStopping(
    monitor='val_accuracy',        # نراقب خسارة التحقق
    patience=10,                # لو مفيش تحسن 5 مرات متتالية → نوقف التدريب
    verbose=1,                 # إظهار رسالة عند التوقف
    restore_best_weights=True  # نرجع الأوزان لأفضل نسخة تم حفظها
)

# --------------------------------------

# 3️⃣ ReduceLROnPlateau → تقليل معدل التعلم (Learning Rate) تلقائيًا
reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',    # نراقب خسارة التحقق
    factor=0.3,            # نقلل معدل التعلم للنصف (lr = lr * 0.5)
    patience=2,            # لو مفيش تحسن لمدة 3 مرات → نطبّق التغيير
    verbose=1,             # إظهار رسالة عند تغيير الـ lr
    min_lr=1e-8            # أقل معدل تعلم ممكن نوصل له
)

# --------------------------------------

# 4️⃣ دمج كل الـ callbacks في قائمة واحدة
callbacks_list = [checkpoint, earlystop, reduce_lr]
```

```python
# ==========================
# 7️⃣ تدريب النموذج
# ==========================
history = model.fit(
    train_generator,            # بيانات التدريب
    validation_data=val_generator,  # بيانات التحقق
    epochs=50,                  # عدد epochs كبير (EarlyStopping سيوقف التدريب عند الحاجة)
    class_weight=class_weights, # توازن الفئات
    callbacks=callbacks_list    # إضافة callbacks
)
```

```python
# فتح بعض الطبقات للتدريب
base_model.trainable = True
for layer in base_model.layers[:-30]:
    layer.trainable = False

# إعادة تجميع النموذج بمعدل تعلم أصغر
model.compile(optimizer=Adam(learning_rate=1e-5),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# إعادة التدريب
history_finetune = model.fit(
    train_generator,
    epochs=15,
    validation_data=val_generator,
    class_weight=class_weights,
    callbacks=callbacks_list
)
```

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# ==========================
# 1️⃣ تحميل النموذج
# ==========================
model = load_model("best_skin_model_1_Dremax.keras")
print("✅ Model loaded successfully!")

# ==========================
# 2️⃣ تجهيز بيانات الاختبار
# ==========================
IMG_SIZE = (224, 224)   # نفس حجم التدريب
BATCH_SIZE = 32

test_dir = "/content/dataset/test"

test_datagen = ImageDataGenerator(rescale=1./255)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False   # مهم عشان التقييم يطلع مضبوط
)

# ==========================
# 3️⃣ تقييم النموذج
# ==========================
loss, accuracy = model.evaluate(test_generator)
print(f"📊 Test Loss: {loss:.4f}")
print(f"✅ Test Accuracy: {accuracy:.4f}")

# ==========================
# 4️⃣ تقرير تفصيلي
# ==========================
y_pred = model.predict(test_generator)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = test_generator.classes
class_labels = list(test_generator.class_indices.keys())

print("\n📋 Classification Report:")
print(classification_report(y_true, y_pred_classes, target_names=class_labels))

# ==========================
# 5️⃣ مصفوفة الالتباس (Heatmap)
# ==========================
cm = confusion_matrix(y_true, y_pred_classes)

plt.figure(figsize=(12,8))
sns.heatmap(cm, annot=False, cmap="Blues", xticklabels=class_labels, yticklabels=class_labels)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()
```
